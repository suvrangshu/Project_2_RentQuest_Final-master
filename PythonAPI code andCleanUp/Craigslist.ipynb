```python
from craigslist import CraigslistHousing
import pandas as pd
import os
import csv
import dateutil
import time
import json
import numpy as np
from pygeocoder import Geocoder

from citipy import citipy

import warnings
warnings.filterwarnings('ignore')


    
```

```python
tmppd = []
#output_data_file1 = "data/craiglistMarch.csv"
```

```python
#sfc, sby, eby,pen,nby 
#cl_h = CraigslistHousing(site='sfbay', area='sfc', category='apt', \
   #                      filters={'max_price': 1200, 'private_room': True})
region = ['sfc', 'sby', 'eby', 'nby', 'pen']

for xloc in region:
    print("Now looking in: ", xloc)
    region_code = xloc
    cl_h = CraigslistHousing(site='sfbay', area=xloc, category='apa',\
                         filters={'max_price': 4000, 'availabilityMode': 1})
    i = 0
    for result in cl_h.get_results(sort_by='newest', geotagged=True):
       # i += 1
        result.update({'Region': region_code}) # add region code
        tmppd1 = dict(result)
        tmppd.append(tmppd1)
       # if i == 10:
        #    break
    pd1 = pd.DataFrame(tmppd)
  
    
#call city py with lat , lon to get city location
for index, row in pd1.iterrows():
    #pd1.loc[index,"Region"] = region_code
        #pd1.loc[index,"Region"] = loc
    if not row['geotag'] is None:
        xtemp1 = row['geotag'][0]
        xtemp2 = row['geotag'][1]
        pd1.loc[index,"Latitude"] = xtemp1
        pd1.loc[index,"Longtitude"] = xtemp2
        city = citipy.nearest_city(xtemp1, xtemp2).city_name
        pd1.loc[index,"City"] = city
        craigpd = pd1
        # check bedroom 
        if row['bedrooms'] is None: # empty bedroom = studio
            #print('empty bed: ',row['bedrooms'])
            pd1.loc[index,"bedrooms"] = 0
    else:
        print('geotag empty') # skip the record
        





```
