```python
#group by city, count total number of rentals for all cities

city_rental = mydb.craigslist_collection.aggregate([
    { "$group": {
        "_id": "$City",
        "count": { "$sum": 1 }}}
      
    
])


list(city_rental)

```

```python
# city Average rental by Region


regional_rental_avg = mydb.craigslist_collection.aggregate([
    
    {
        "$group": {
            "_id":"$Region", 
            "Reg_avg_price": {"$avg": "$price"}
        }
    }
      
    
])

list(regional_rental_avg)
        
```

```python
#City rental average price all cities
city_rental_avg = mydb.craigslist_collection.aggregate([
    
    {
        "$group": {
            "_id":"$City", 
            "avg_price": {"$avg": "$price"}
        }
    }
      
    
])

list(city_rental_avg)
```

```python
#City rental Max price all cities
city_rental_max = mydb.craigslist_collection.aggregate([
    
    {
        "$group": {
            "_id":"$City", 
            "max_price": {"$max": "$price"}
        }
    }
      
    
])

list(city_rental_max)
```

```python
#max price by individual city:
xCity = 'alameda'

city_rental_max = mydb.craigslist_collection.aggregate([
    
    {
        "$match" : {"City":xCity}
    },
    {
        "$group" : {
            "_id": "$City", 
            "max_rent" : {"$max" : "$price"}
        }
    }

])

list(city_rental_max)
```

```python
#min rental price by individual city:
xCity = 'alameda'

city_rental_max = mydb.craigslist_collection.aggregate([
    
    {
        "$match" : {"City":xCity}
    },
    {
        "$group" : {
            "_id": "$City", 
            "max_rent" : {"$min" : "$price"}
        }
    }

])

list(city_rental_max)
```

```python
#total rental count by city = 'alameda'

xCity = 'alameda'
tot_rental_by_one_city = mydb.craigslist_collection.aggregate([
    {
        "$match" : {"City":xCity}
    },
    
    {
        "$group": {
            "_id":"$City", 
            "Total_rental": {"$sum": 1}
        }
    }
      
    
])

list(tot_rental_by_one_city)
        
```
